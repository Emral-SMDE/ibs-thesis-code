## 📊 Data Loading & Initial Inspection

This is the first step before we start EDA. Here, we begin by importing the dataset and performing basic checks to understand its structure and quality.

We load the data into a DataFrame, then check the shape, column names, and data types. We also display the first few rows to get an overview of the content. We examine data quality by checking for missing values and duplicate rows, which could impact further analysis.

Finally, we generate descriptive statistics to summarize the numeric and categorical features, helping us to understand distributions.

```python
# Data Loading & Initial Inspection

import pandas as pd

# Loading the dataset
df = pd.read_csv('heart_attack_china.csv')

# Displays basic information
print(f"Dataset Shape: {df.shape}")
print("\nColumn Names:")
print(df.columns.tolist())
print("\nData Types:")
print(df.dtypes)

# Shows first few rows
print("\nFirst 5 Rows:")
display(df.head())

# Checks for missing values
print("\nMissing Values (per column):")
missing = df.isnull().sum()
display(missing)

# Checks for duplicate rows
print(f"\nNumber of duplicate rows: {df.duplicated().sum()}")

# Descriptive statistics
print("\nDescriptive Statistics (Numeric):")
display(df.describe())

print("\nDescriptive Statistics (All Columns):")
display(df.describe(include='all'))
```

## 📊 Univariate Analysis — Categorical Features

Here, we analyze the distribution of each categorical feature in the dataset.

We first identify all categorical columns. Then, for each of these columns, we:

- Count how often each category appears (including missing values).
- Visualize these counts using a bar chart.
- Add value labels on top of each bar for clarity.
- Customize the plots for readability.

This helps us understand the frequency distribution of each categorical variable and identify any imbalances or unusual patterns in the data.


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


categorical_cols = df.select_dtypes(include=['object']).columns.tolist()

for col in categorical_cols:
    counts = df[col].value_counts(dropna=False)
    labels = counts.index.astype(str)
    values = counts.values

    colors = sns.color_palette('Set2', n_colors=len(labels))

    plt.figure(figsize=(8, 5))
    bars = plt.bar(labels, values, color=colors)

    # Annotate
    for bar in bars:
        h = bar.get_height()
        if h > 0:
            plt.text(
                bar.get_x() + bar.get_width()/2, h,
                f'{int(h)}',
                ha='center', va='bottom'
            )

    plt.title(f'Value Counts of {col}')
    plt.xlabel(col.replace('_',' '))
    plt.ylabel('Count')
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.show()
```

## 📊 Bivariate Analysis — Categorical Features vs. Heart Attack

Here, we examine how each categorical feature is related to the target variable `Heart_Attack`.

We exclude the target column from the list of categorical features, then create count plots for each feature, grouped by heart attack status.

Each plot shows how the frequency of different categories varies between people who had a heart attack and those who did not. Annotations are added on top of the bars for better readability.

This helps us to explore potential relationships between categorical variables (like gender, smoking status, etc.) and the likelihood of heart attacks.


```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Get all categorical columns except the target
categorical_cols = df.select_dtypes(include=['object']).columns.tolist()
categorical_cols.remove('Heart_Attack')

# Loop through each feature
for col in categorical_cols:
    plt.figure(figsize=(8, 5))
    ax = sns.countplot(data=df, x=col, hue='Heart_Attack', palette='Set2')

    # Annotate only non-zero bars
    for p in ax.patches:
        height = p.get_height()
        if height > 0:
            ax.annotate(
                f'{int(height)}',
                (p.get_x() + p.get_width() / 2, height),
                ha='center', va='bottom'
            )

    ax.set_title(f'Heart Attack Counts by {col.replace("_", " ")}')
    ax.set_ylabel('Frequency')
    ax.set_xlabel(col.replace('_', ' '))
    plt.legend(title='Heart Attack')
    plt.tight_layout()
    plt.show()
```

## 🧠 Data Encoding for Machine Learning

Before applying machine learning models, we need to convert categorical values into numerical format.

Here, we:
- **Drop irrelevant features** that won’t be used in modeling (e.g., ID, location, income).
- **Categorize features** into four types:
  - **Numeric**: left unchanged.
  - **Ordinal**: encoded with custom order (e.g., Low < Medium < High).
  - **Binary**: encoded as 0/1 using specific value mapping.
  - **One-hot**: used for nominal categories (like Gender).

We build a `ColumnTransformer` to apply the appropriate encoding to each group. This is wrapped in a `Pipeline` for easy transformation.

Finally, we split the dataset into features `X` and target `y`, encode the features using the pipeline, and check the shape and type of the resulting encoded matrix.

This prepares the data in a clean numerical format ready for model training.


```python
import pandas as pd
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OneHotEncoder, OrdinalEncoder


# Define drop, numeric, ordinal, binary, onehot features
drop_features = [
    'Patient_ID', 'Region', 'Province',
    'Rural_or_Urban', 'Employment_Status',
    'Education_Level', 'Income_Level'
]
numeric_features = ['Age', 'Blood_Pressure', 'CVD_Risk_Score']
ordinal_features = [
    'Cholesterol_Level', 'Air_Pollution_Exposure',
    'Physical_Activity', 'Diet_Score',
    'Stress_Level', 'Healthcare_Access',
    'Hospital_Availability'
]
binary_features = [
    'Smoking_Status', 'Alcohol_Consumption', 'Diabetes',
    'Hypertension', 'Obesity', 'Family_History_CVD',
    'Chronic_Kidney_Disease', 'Previous_Heart_Attack',
    'TCM_Use'
]
onehot_features = ['Gender']

# Ordinal categories
ordinal_categories = [
    ['Low', 'Normal', 'High'],
    ['Low', 'Medium', 'High'],
    ['Low', 'Medium', 'High'],
    ['Poor', 'Moderate', 'Healthy'],
    ['Low', 'Medium', 'High'],
    ['Poor', 'Moderate', 'Good'],
    ['Low', 'Medium', 'High']
]

# Accurate binary categories per column
binary_categories = [
    ['Non-Smoker', 'Smoker'],  # Smoking_Status
    ['No', 'Yes'],             # Alcohol_Consumption
    ['No', 'Yes'],             # Diabetes
    ['No', 'Yes'],             # Hypertension
    ['No', 'Yes'],             # Obesity
    ['No', 'Yes'],             # Family_History_CVD
    ['No', 'Yes'],             # Chronic_Kidney_Disease
    ['No', 'Yes'],             # Previous_Heart_Attack
    ['No', 'Yes']              # TCM_Use
]

# Build preprocessor
preprocessor = ColumnTransformer(
    transformers=[
        ('ord', OrdinalEncoder(categories=ordinal_categories), ordinal_features),
        ('bin', OrdinalEncoder(categories=binary_categories), binary_features),
        ('ohe', OneHotEncoder(handle_unknown='ignore'), onehot_features),
        ('num', 'passthrough', numeric_features)
    ],
    remainder='drop'
)

pipeline = Pipeline([
    ('preprocessor', preprocessor)
])

# Prepare X and y
X = df.drop(columns=drop_features + ['Heart_Attack'])
y = df['Heart_Attack'].map({'No': 0, 'Yes': 1})

# Fit and transform
X_encoded = pipeline.fit_transform(X)

# Display result
print("Encoded matrix shape:", X_encoded.shape)
print("Encoded matrix type:", type(X_encoded))
```

```python


# 1. Get the output feature names
feature_names = pipeline.named_steps['preprocessor'].get_feature_names_out()

# 2. Create a DataFrame with the encoded data
encoded_df = pd.DataFrame(X_encoded, columns=feature_names)

# 3. Inspect the first few rows
encoded_df.head()
```

```python
# 1. Transform X
X = df.drop(columns=drop_features + ['Heart_Attack'])
y_ha = df['Heart_Attack'].map({'No': 0, 'Yes': 1})
y_cvd = df['CVD_Risk_Score']

# fit_transform if not yet fitted, or just transform:
X_enc = pipeline.fit_transform(X)

# 2. Recover feature names
pre = pipeline.named_steps['preprocessor']
feat_names = pre.get_feature_names_out()
X_enc = pd.DataFrame(X_enc, columns=feat_names, index=df.index)

# 3. Define thematic groups by name pattern
clinical_feats = [f for f in feat_names if f in ('num__Age', 'num__Blood_Pressure')]
binary_feats  = [f for f in feat_names if f.startswith('bin__')]
ordinal_feats = [f for f in feat_names if f.startswith('ord__')]
gender_feats  = [f for f in feat_names if f.startswith('ohe__Gender_')]

groups = {
    'Clinical': clinical_feats,
    'Binary Flags': binary_feats,
    'Ordinals': ordinal_feats,
    'Gender (one‐hot)': gender_feats
}

# 4. Compute & print correlations
for name, feats in groups.items():
    corr_ha  = X_enc[feats].corrwith(y_ha).round(5)
    corr_cvd = X_enc[feats].corrwith(y_cvd).round(5)
    table = pd.concat([corr_ha, corr_cvd], axis=1)
    table.columns = ['corr_with_Heart_Attack','corr_with_CVD_Risk_Score']
    print(f"\n=== {name} ===")
    display(table)
```

```python
# Plot heatmaps for each group
for name, feats in groups.items():
    # build correlation table
    corr_df = pd.concat([
        X_enc[feats].corrwith(y_ha),
        X_enc[feats].corrwith(y_cvd)
    ], axis=1).round(5)
    corr_df.columns = ['Heart_Attack', 'CVD_Risk_Score']

    # plot
    plt.figure(figsize=(6, len(feats)*0.5 + 1))
    sns.heatmap(corr_df, annot=True, fmt='.5f', cmap='coolwarm', vmin=-1, vmax=1, cbar_kws={'shrink': .8})
    plt.title(f'Correlation Matrix: {name}', pad=10)
    plt.ylabel('')
    plt.tight_layout()
    plt.show()
```

```python
import warnings
import numpy as np
import pandas as pd

from sklearn.model_selection import train_test_split, StratifiedKFold, cross_validate
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from xgboost import XGBClassifier
from lightgbm import LGBMClassifier
from catboost import CatBoostClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import make_scorer, precision_score, roc_auc_score, f1_score, classification_report

# Silence warnings
warnings.filterwarnings("ignore", category=FutureWarning)
warnings.filterwarnings("ignore", category=UserWarning)

# 1. Train/test split
X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.2,
    stratify=y,
    random_state=42
)

# 2. Custom precision scorer to avoid undefined warnings
precision_scorer = make_scorer(precision_score, zero_division=0)

# 3. Define models (excluding SVM)
models = [
    ('LogisticRegression',  LogisticRegression(class_weight='balanced', max_iter=1000, random_state=42)),
    ('RandomForest',        RandomForestClassifier(class_weight='balanced', n_jobs=1, random_state=42)),
    ('GradientBoosting',    GradientBoostingClassifier(n_estimators=100, random_state=42)),
    ('XGBoost',             XGBClassifier(eval_metric='logloss', n_estimators=50, n_jobs=1, random_state=42)),
    ('LightGBM',            LGBMClassifier(class_weight='balanced', n_estimators=50, n_jobs=1, random_state=42)),
    ('CatBoost',            CatBoostClassifier(iterations=100, verbose=0, random_state=42)),
    ('KNN',                 KNeighborsClassifier(n_jobs=1))
]

# 4. Cross-validation on the TRAIN set
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
scoring = {
    'Accuracy':  'accuracy',
    'ROC_AUC':   'roc_auc',
    'Precision': precision_scorer,
    'Recall':    'recall',
    'F1':        'f1'
}

results = []
for name, clf in models:
    pipe = Pipeline([
        ('preprocessor', preprocessor),
        ('classifier',    clf)
    ])
    cv_res = cross_validate(
        pipe, X_train, y_train,
        cv=cv,
        scoring=scoring,
        n_jobs=-1,
        return_train_score=False
    )
    scores = {metric: np.mean(cv_res[f'test_{metric}']) for metric in scoring}
    scores['Model'] = name
    results.append(scores)

# 5. Compile and display CV results
results_df = pd.DataFrame(results).set_index('Model')
results_df = results_df[['Accuracy','ROC_AUC','Precision','Recall','F1']].round(4)
print(results_df)

# 6. Final evaluation on TEST set using RandomForest
best_pipe = Pipeline([
    ('preprocessor', preprocessor),
    ('classifier',    RandomForestClassifier(class_weight='balanced', random_state=42))
])
best_pipe.fit(X_train, y_train)
y_pred = best_pipe.predict(X_test)
y_proba = best_pipe.predict_proba(X_test)[:, 1]

test_roc = roc_auc_score(y_test, y_proba)
test_f1  = f1_score(y_test, y_pred, zero_division=0)

print("\nFinal Test Metrics")
print("------------------")
print("ROC AUC:     ", round(test_roc, 4))
print("F1 Score:   ", round(test_f1, 4))
print(classification_report(y_test, y_pred, zero_division=0))
```

```python
# 1. Best model *per* metric:
best_per_metric = results_df.idxmax()
print("Best model by metric:")
print(best_per_metric, "\n")

# 2. Compute a “mean rank” across all metrics:
#    (rank=1 for highest value, 2 for second, etc.)
rank_df = results_df.rank(ascending=False, method='min')
rank_df['mean_rank'] = rank_df.mean(axis=1)

print("Rankings by metric, and overall mean rank:")
print(rank_df.sort_values('mean_rank'))

# 3. Overall winner:
overall = rank_df['mean_rank'].idxmin()
print(f"\n>>> Overall best model (lowest mean rank): {overall}")
```

```python
from imblearn.over_sampling import SMOTE
from imblearn.pipeline import Pipeline as ImbPipeline
import numpy as np
import pandas as pd

# 2. Define SMOTE step
smote = SMOTE(random_state=42)

# 3. Evaluate each model with SMOTE in the pipeline
smote_results = []
for name, clf in models:
    imb_pipe = ImbPipeline([
        ('preprocessor', preprocessor),
        ('smote',        smote),
        ('classifier',   clf)
    ])
    cv_res = cross_validate(
        imb_pipe,
        X_train, y_train,
        cv=cv,
        scoring=scoring,
        n_jobs=-1,
        return_train_score=False
    )
    scores = {metric: np.mean(cv_res[f'test_{metric}']) for metric in scoring}
    scores['Model'] = name
    smote_results.append(scores)

# 4. Tabulate
smote_df = pd.DataFrame(smote_results).set_index('Model')
smote_df = smote_df[['Accuracy','ROC_AUC','Precision','Recall','F1']].round(4)

print("=== CV RESULTS WITH SMOTE ===")
print(smote_df)
```
